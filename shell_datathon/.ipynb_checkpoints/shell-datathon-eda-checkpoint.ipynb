```python
sns.distplot(df_train['Litres'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df_train['Litres'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('Litres values distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(df_train['Litres'], plot=plt)
plt.show()
```

# Group by 

```python
# Create Date Columns 
df_train["Billing Date"] = pd.to_datetime(df_train["Billing Date"])
df_train["Year"] = df_train["Billing Date"].dt.year
df_train["Month"] = df_train["Billing Date"].dt.month
df_train["Week"] = df_train["Billing Date"].dt.week
```

```python
table_2 = df_train.groupby(["Year"])["Litres"].agg([np.mean, np.sum, np.max, np.min, np.size])
pd.DataFrame(table_2).style.background_gradient(cmap = "winter")
```

```python
fig = px.bar(pd.DataFrame(table_2), x=pd.DataFrame(table_2).index, y='sum',
            color_discrete_sequence =['blue']*len(table_2),
             title="Total Litre values grouping by Year")
fig.show()
```

# graphics explain a lot of information :)

```python
table_3 = df_train.groupby(["Area"])["Litres"].agg([np.mean, np.sum, np.max, np.min, np.size])
pd.DataFrame(table_3).style.background_gradient(cmap = "Paired")
```

```python
fig = px.bar(pd.DataFrame(table_3), x=pd.DataFrame(table_3).index, y='sum',
            color_discrete_sequence =['red']*len(table_3),
             title="Total Litre values grouping by Area")
fig.show()
```

```python

label_encoder = preprocessing.LabelEncoder()
df_train['City Code']= label_encoder.fit_transform(df_train['City Code'])
  
table_4 = df_train.groupby(["City Code"])["Litres"].agg([np.mean, np.sum, np.max, np.min, np.size])
pd.DataFrame(table_4).style.background_gradient(cmap = "Paired")
```

```python
fig = px.bar(pd.DataFrame(table_4), x=pd.DataFrame(table_4).index, y='sum',
             color_discrete_sequence =['green']*len(table_4),
             title="Total Litre values grouping by city")
fig.show()
```

```python
table_5 = df_train.groupby(["Product"])["Litres"].agg([np.mean, np.sum, np.max, np.min, np.size])
pd.DataFrame(table_5).style.background_gradient(cmap = "Paired")
```

```python
fig = px.bar(pd.DataFrame(table_5), x=pd.DataFrame(table_5).index, y='sum',
             color_discrete_sequence =['black']*len(table_5),
             title="Total Litre values grouping by Product Type")
fig.show()
```

# Missing Values that we don't have

```python
miss = df_train.isna().sum().sort_values(ascending=False).reset_index(drop=False).rename({0:'missing_count'}, axis=1)
miss['miss_ratio'] = (miss['missing_count'] / df_train.shape[0]) * 100

plt.figure(figsize= (18,12))
sns.barplot(data=miss[:25], x= 'miss_ratio', y='index', palette='YlOrBr_r', linewidth=0.7, edgecolor=".2")
plt.title('Missing Value Ratios Per Column')
plt.ylabel('Column')
plt.xlabel('Percent of Missing Values')
del miss
plt.show()
```
