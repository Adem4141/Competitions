```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)

# test setindeki idlerin bazıları train setinde yok onları ayrı değerlendirmek gerekiyor
```

```python
price = pd.read_csv("price.csv")
sample_submission = pd.read_csv("sample_submission.csv")
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv").drop("index",axis=1)
```

```python
for df in [price, sample_submission, train, test]:
    df.columns = [i.replace(" ", "").lower() for i in df.columns]
```

```python
train.shape,test.shape, sample_submission.shape
```

```python
train["billingdate"]=pd.to_datetime(train["billingdate"])
test["weekstartdate"]=pd.to_datetime(test["weekstartdate"])
price["validfrom"]=pd.to_datetime(price["validfrom"])
```

```python
train = pd.merge(train, price, left_on = ["citycode","billingdate","product"], right_on=["citycode","validfrom","product"], how="left")
```

```python
train["week_of_year"]=train["billingdate"].dt.isocalendar().week
train["year"] = train["billingdate"].dt.year
test=pd.merge(test, price, left_on = ["citycode","weekstartdate","product"], right_on=["citycode","validfrom","product"], how="left")
```

```python
agg_train = train.groupby(["customer","area","citycode","year","week_of_year","product"]).agg({"litres":"sum","crp":"mean"}).reset_index()
```

```python
# Logaritmik dönüşüm
gercek_log = np.log1p(y_test)
tahmin_log = np.log1p(preds)

# Karelerin farkı
karelerin_farki = (tahmin_log - gercek_log) ** 2

# Ortalama alın
mse = np.mean(karelerin_farki)

# Kökünü alarak RMSLE hesapla
rmsle = np.sqrt(mse)

```
